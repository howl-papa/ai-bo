```python
import PyPDF2

# PDF 파일 열기
with open(r"C:\Users\USER\Desktop\MS_2nd_project\현대해상_굿앤굿어린이종합보험.pdf", "rb") as file:
    reader = PyPDF2.PdfReader(file)
    total_pages = len(reader.pages)

    print(f"총 페이지 수: {total_pages}")

    # 전체 텍스트 모으기
    all_text = ""
    for i, page in enumerate(reader.pages):
        text = page.extract_text()
        print(f"페이지 {i+1} 추출 완료")
        all_text += text + "\n"

# 추출된 텍스트 저장하기
with open("extracted_text.txt", "w", encoding="utf-8") as f:
    f.write(all_text)
```

```python
import PyPDF2
import os

def extract_text_from_pdf(pdf_path):
    """
    PDF 파일에서 텍스트를 추출하여 반환합니다.

    Args:
        pdf_path (str): 텍스트를 추출할 PDF 파일의 경로.

    Returns:
        str: 추출된 PDF 텍스트. 오류 발생 시 빈 문자열을 반환합니다.
    """
    extracted_text = ""
    if not os.path.exists(pdf_path):
        print(f"오류: 파일을 찾을 수 없습니다 - {pdf_path}")
        return ""

    try:
        # PDF 파일을 바이너리 읽기 모드로 엽니다.
        with open(pdf_path, 'rb') as file:
            # PdfReader 객체를 생성합니다.
            reader = PyPDF2.PdfReader(file)

            # PDF의 각 페이지를 순회하며 텍스트를 추출합니다.
            for page_num in range(len(reader.pages)):
                page = reader.pages[page_num]
                extracted_text += page.extract_text() + "\n" # 각 페이지의 텍스트와 줄바꿈 추가

    except PyPDF2.errors.PdfReadError:
        print(f"오류: 유효한 PDF 파일이 아니거나 손상되었습니다 - {pdf_path}")
    except Exception as e:
        print(f"텍스트 추출 중 오류 발생: {e}")

    return extracted_text

# --- 사용 예시 ---
if __name__ == "__main__":
    # 여기에 실제 PDF 파일 경로를 입력하세요.
    # 예시: 'document.pdf' 또는 'C:/Users/YourUser/Documents/report.pdf'
    pdf_file_path = "현대해상약관\현대해상_굿앤굿어린이종합보험.pdf" # 이 부분을 실제 PDF 파일 경로로 변경해주세요.

    # 텍스트 추출 함수 호출
    text_content = extract_text_from_pdf(pdf_file_path)

    if text_content:
        print("--- PDF에서 추출된 텍스트 ---")
        print(text_content[:1000]) # 추출된 텍스트의 처음 1000자만 출력 (너무 길 경우를 대비)
        print("\n--- 전체 텍스트가 추출되었습니다. ---")
        # 추출된 전체 텍스트를 파일로 저장하고 싶다면 아래 주석을 해제하세요.
        with open("extracted_text.txt", "w", encoding="utf-8") as f:
            f.write(text_content)
        print("추출된 텍스트가 'extracted_text.txt' 파일로 저장되었습니다.")
    else:
        print("텍스트를 추출하지 못했습니다.")

```

```python
import re
import pandas as pd

# 조항 추출 함수 정의
def extract_article_info(text):
    if pd.isnull(text):
        return None, None

    # 조항 패턴 탐색
    match = re.search(r'(제\d+조)(?:\s*\(?([^)]+)\)?)?', text)
    if match:
        article_number = match.group(1)  # 예: 제5조
        article_title = match.group(2) if match.group(2) else ""  # 예: 보험금의 지급사유
        return article_number, article_title
    else:
        return None, None
```

```python
# 앞서 만든 df_split을 불러왔다고 가정
df_split = pd.read_csv("chunks_split_korean_final.csv")

# 조항 정보 추출하여 새 열 추가
df_split[['article_number', 'article_title']] = df_split['text'].apply(
    lambda x: pd.Series(extract_article_info(x))
)

# 결과 저장
df_split.to_csv("chunks_with_articles.csv", index=False, encoding="utf-8-sig")

# 미리보기
print(df_split[df_split['article_number'].notnull()].head(10))
```
